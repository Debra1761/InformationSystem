```python
import requests 
api_key="bedd7e041002427cbc44ffb3a1fefa94"

r = requests.get("https://newsapi.org/v2/top-headlines?country=us&apiKey={}".format(api_key))
```

```python
article_urls = r.json()
```

```python
urls = []
for article in article_urls['articles']:
    urls.append(article['url'])
    
```

```python
from newspaper import Article
```

```python
article_texts = []

for url in urls:
    article = Article(url)
    article.download()
    article.parse()
    author = article.authors[0] if len(article.authors)>0 else ''
    publish_date = article.publish_date
    print(publish_date)
    article_texts.append(article.text)
#     print(article.text)
```

```python
article_texts
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
entities = []
for text in article_texts: 
    doc = nlp(text)

    for ent in doc.ents:
        entities.append(ent.text)
        print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
from collections import Counter

Counter(entities).most_common()
```
